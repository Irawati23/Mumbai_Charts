```python
mumleave= pd.read_csv('/Users/Irawati/Desktop/mumleave.csv')
mumleave['district_eng']=mumleave['district_eng'].replace(['Mumbai'], 'Mumbai city')
mumleave
```

```python
DISTRICT_ENG=mumleave['district_eng']
COUNT=mumleave['Count']
YEAR= mumleave['year']
figmum= px.treemap(mumleave,path=[DISTRICT_ENG,YEAR],values=COUNT, color=COUNT, color_continuous_scale=['green','yellow','red'],title='Leave and license agreement registrations by district from 2011 to 2020')
figmum.update_layout(title_font_size=40,title_font_family= 'Arial')
figmum.layout.font.size = 18

plotly.offline.plot(figmum, filename='Yearly leave and licence registrations trend by district.html')
```
